```python
# Criação de uma feature para 'Transação de Baixo Valor'

# Usar um percentil de fraudes como limiar
fraud_amount_75th_percentile = df[df['Class'] == 1]['Amount'].quantile(0.75)

low_amount_threshold = fraud_amount_75th_percentile


df['Is_Low_Amount'] = (df['Amount'] <= low_amount_threshold).astype(int)
print(f"Nova coluna 'Is_Low_Amount' criada com limiar de {low_amount_threshold:.2f}: {df['Is_Low_Amount'].head()}")
print(f"Proporção de 'Is_Low_Amount':\n{df['Is_Low_Amount'].value_counts(normalize=True).round(2)}%")

```

###  Insight Distribuição de Transações por Hora

A análise inicial do volume de transações por hora confirma um padrão esperado de atividade diária: **baixo volume na madrugada (1h-6h)** e aumento gradual a partir do amanhecer.  

No entanto, **devido ao desbalanceamento dos dados**, essa visualização não revela o comportamento das transações fraudulentas (Classe 1).  

Para entender a fraude, é **crucial analisar a distribuição da classe minoritária separadamente**, buscando identificar se, mesmo em horários de baixo volume geral, a **proporção ou o número absoluto de fraudes é desproporcionalmente alto**.  

Isso pode indicar **possíveis janelas de oportunidade exploradas por golpistas**, o que pode ser essencial para a construção de regras de detecção ou definição de alertas em tempo real.


```python
# Calcular a Proporção de Fraudes por Hora 

# Agrupar por 'Hour' e 'Class', contar as transações
grouped_by_hour_class = df.groupby(['Hour', 'Class']).size().unstack(fill_value=0)

# Calcular a proporção de fraudes (Classe 1)
grouped_by_hour_class['Total'] = grouped_by_hour_class[0] + grouped_by_hour_class[1]
grouped_by_hour_class['Fraude_Proporcao'] = (grouped_by_hour_class[1] / grouped_by_hour_class['Total']) * 100

# Reiniciar o índice para que 'Hour' seja uma coluna novamente
df_plot = grouped_by_hour_class.reset_index()

# Plotar o Gráfico de Linhas da Proporção de Fraudes
plt.figure(figsize=(12, 6))
sns.lineplot(x='Hour', y='Fraude_Proporcao', data=df_plot, marker='o', color='red')
plt.title('Porcentagem de Transações Fraudulentas por Hora do Dia')
plt.xlabel('Hora do Dia')
plt.ylabel('Porcentagem de Fraudes (%)')
plt.xticks(ticks=df_plot['Hour'], rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

###  Insight: Padrões Horários nas Transações Fraudulentas

A análise da **proporção de fraudes ao longo das horas do dia** revela **comportamentos atípicos durante a madrugada**, com **picos significativos de fraudes entre 1h e 3h da manhã**, especialmente por volta das **2h**, onde a taxa de fraudes ultrapassa **1,7% das transações realizadas**.

Esses picos ocorrem justamente em um período de **baixo volume geral de transações**, conforme identificado na análise anterior, o que **reforça a hipótese de que fraudadores exploram horários de menor vigilância** ou menor tráfego no sistema para aplicar golpes.

Além disso, observa-se que durante o restante do dia (especialmente entre 6h e 23h), a **proporção de fraudes se mantém consistentemente abaixo de 0,5%**, indicando um comportamento menos agressivo ou maior detecção nesses períodos.

✅ **Conclusão:** Os resultados sugerem a existência de **"janelas de oportunidade" entre 1h e 3h**, nas quais é recomendável intensificar a **monitoria automática** e implementar **regras de negócio mais rígidas** para reduzir os riscos de fraude.

```python
# Escolha de um modelo (vamos começar com Regressão Logística e Random Forest)

# Modelo 1: Regressão Logística
print("\n--- Treinando Regressão Logística ---")
log_reg = LogisticRegression(solver='liblinear', random_state=42)
# Se você usou SMOTE, não precisa de class_weight='balanced'
#log_reg = LogisticRegression(solver='liblinear', random_state=42)

log_reg.fit(X_train_resampled, y_train_resampled)
y_pred_log_reg = log_reg.predict(X_test)
y_prob_log_reg = log_reg.predict_proba(X_test)[:, 1]

print("\nRelatório de Classificação (Regressão Logística):")
print(classification_report(y_test, y_pred_log_reg))

print("\nMatriz de Confusão (Regressão Logística):")
print(confusion_matrix(y_test, y_pred_log_reg))

auc_log_reg = roc_auc_score(y_test, y_prob_log_reg)
print(f"AUC-ROC (Regressão Logística): {auc_log_reg:.4f}")
```

```python
# Modelo 2: Random Forest
print("\n--- Treinando Random Forest ---")
# N_estimators e max_depth são hiperparâmetros que você pode ajustar.
# class_weight='balanced' é uma alternativa ao SMOTE se não quiser reamostrar.
rf_clf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, class_weight='balanced')
# Se você usou SMOTE, não precisa de class_weight='balanced'
# rf_clf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

rf_clf.fit(X_train_resampled, y_train_resampled)
y_pred_rf = rf_clf.predict(X_test)
y_prob_rf = rf_clf.predict_proba(X_test)[:, 1]

print("\nRelatório de Classificação (Random Forest):")
print(classification_report(y_test, y_pred_rf))

print("\nMatriz de Confusão (Random Forest):")
print(confusion_matrix(y_test, y_pred_rf))

auc_rf = roc_auc_score(y_test, y_prob_rf)
print(f"AUC-ROC (Random Forest): {auc_rf:.4f}")

# Visualizar Curva ROC
plt.figure(figsize=(8, 6))
fpr_log_reg, tpr_log_reg, _ = roc_curve(y_test, y_prob_log_reg)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)

plt.plot(fpr_log_reg, tpr_log_reg, label=f'Regressão Logística (AUC = {auc_log_reg:.2f})')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {auc_rf:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Aleatório')
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva ROC')
plt.legend()
plt.grid()
plt.show()
```
